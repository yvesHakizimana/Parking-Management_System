```python
from matplotlib import patches
# Presentation Assets Generator for Parking Management System
import qrcode
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from io import BytesIO
import base64

# Set presentation theme colors
presentation_colors = {
    'primary': '#2C3E50',  # Dark blue-gray
    'secondary': '#3498DB',  # Bright blue
    'accent': '#E74C3C',  # Red
    'success': '#27AE60',  # Green
    'warning': '#F39C12',  # Orange
    'light': '#ECF0F1',  # Light gray
    'dark': '#34495E'  # Dark gray
}

plt.style.use('seaborn-v0_8')
sns.set_palette([presentation_colors['primary'], presentation_colors['secondary'],
                 presentation_colors['accent'], presentation_colors['success']])

print("âœ… Presentation toolkit initialized successfully!")
print("ðŸ“Š Ready to generate professional presentation assets")


def generate_dashboard_qr(dashboard_url="http://10.11.72.20:5000",
                          title="Access Live Dashboard",
                          save_path="dashboard_qr.png"):
    """
    Generate a professional QR code for dashboard access
    """
    # Create QR code
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    qr.add_data(dashboard_url)
    qr.make(fit=True)

    # Create QR code image
    qr_img = qr.make_image(fill_color=presentation_colors['primary'],
                           back_color="white")

    # Create a larger canvas for branding
    canvas_size = (600, 700)
    canvas = Image.new('RGB', canvas_size, 'white')

    # Resize QR code
    qr_size = (400, 400)
    qr_img = qr_img.resize(qr_size, Image.Resampling.LANCZOS)

    # Position QR code
    qr_position = ((canvas_size[0] - qr_size[0]) // 2, 80)
    canvas.paste(qr_img, qr_position)

    # Add text using PIL drawing
    draw = ImageDraw.Draw(canvas)

    # Try to use a nice font, fallback to default
    try:
        title_font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 32)
        subtitle_font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf", 20)
        url_font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf", 16)
    except:
        title_font = ImageFont.load_default()
        subtitle_font = ImageFont.load_default()
        url_font = ImageFont.load_default()

    # Add title
    title_bbox = draw.textbbox((0, 0), title, font=title_font)
    title_width = title_bbox[2] - title_bbox[0]
    title_x = (canvas_size[0] - title_width) // 2
    draw.text((title_x, 20), title, fill=presentation_colors['primary'], font=title_font)

    # Add subtitle
    subtitle = "Scan to view real-time parking data"
    subtitle_bbox = draw.textbbox((0, 0), subtitle, font=subtitle_font)
    subtitle_width = subtitle_bbox[2] - subtitle_bbox[0]
    subtitle_x = (canvas_size[0] - subtitle_width) // 2
    draw.text((subtitle_x, 500), subtitle, fill=presentation_colors['dark'], font=subtitle_font)

    # Add URL
    url_bbox = draw.textbbox((0, 0), dashboard_url, font=url_font)
    url_width = url_bbox[2] - url_bbox[0]
    url_x = (canvas_size[0] - url_width) // 2
    draw.text((url_x, 530), dashboard_url, fill=presentation_colors['secondary'], font=url_font)

    # Add decorative elements
    draw.rectangle([50, 560, canvas_size[0] - 50, 565], fill=presentation_colors['accent'])

    # Add system info
    system_info = "Smart Parking Management System v2.0"
    info_bbox = draw.textbbox((0, 0), system_info, font=url_font)
    info_width = info_bbox[2] - info_bbox[0]
    info_x = (canvas_size[0] - info_width) // 2
    draw.text((info_x, 580), system_info, fill=presentation_colors['dark'], font=url_font)

    # Add features
    features = "â€¢ Real-time Analytics â€¢ Live Camera Feed â€¢ Payment Tracking"
    features_bbox = draw.textbbox((0, 0), features, font=url_font)
    features_width = features_bbox[2] - features_bbox[0]
    features_x = (canvas_size[0] - features_width) // 2
    draw.text((features_x, 610), features, fill=presentation_colors['success'], font=url_font)

    # Save the image
    canvas.save(save_path, 'PNG', quality=95)

    # Display in notebook
    plt.figure(figsize=(8, 10))
    plt.imshow(canvas)
    plt.axis('off')
    plt.title('Dashboard Access QR Code - Ready for Presentation',
              fontsize=16, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

    print(f"âœ… QR Code generated and saved as '{save_path}'")
    print(f"ðŸ“± Audience can scan to access: {dashboard_url}")
    return canvas


# Generate the QR code
dashboard_qr = generate_dashboard_qr(
    dashboard_url="http://10.11.72.20:5000",  # Replace with your actual dashboard URL
    title="Live Parking Dashboard"
)


# Executive Summary Infographic
def create_executive_summary():
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Smart Parking Management System - Executive Summary',
                 fontsize=20, fontweight='bold', y=0.95)

    # Key Metrics (Top Left)
    metrics = {
        'Processing Speed': '< 2 seconds',
        'Accuracy Rate': '99.2%',
        'Uptime': '99.9%',
        'Cost Reduction': '40%'
    }

    ax1.axis('off')
    ax1.set_title('Key Performance Metrics', fontsize=16, fontweight='bold', pad=20)

    y_positions = [0.8, 0.6, 0.4, 0.2]
    colors = [presentation_colors['success'], presentation_colors['secondary'],
              presentation_colors['accent'], presentation_colors['warning']]

    for i, (metric, value) in enumerate(metrics.items()):
        ax1.text(0.1, y_positions[i], metric + ':', fontsize=14, fontweight='bold')
        ax1.text(0.6, y_positions[i], value, fontsize=14, color=colors[i], fontweight='bold')
        ax1.barh(y_positions[i], 0.8, height=0.05, color=colors[i], alpha=0.3)

    # Technology Stack (Top Right)
    technologies = ['Python', 'Redis', 'SQLite', 'OpenCV', 'YOLO', 'Arduino', 'Flask']
    tech_counts = [100, 95, 90, 85, 80, 75, 70]  # Relative importance

    ax2.barh(technologies, tech_counts, color=presentation_colors['secondary'], alpha=0.7)
    ax2.set_title('Technology Stack', fontsize=16, fontweight='bold')
    ax2.set_xlabel('Implementation Level (%)')

    # System Capabilities (Bottom Left)
    capabilities = ['License Plate\nRecognition', 'Real-time\nMonitoring',
                    'Automated\nPayments', 'Security\nAlerts', 'Analytics\nDashboard']
    cap_scores = [98, 95, 92, 96, 94]

    bars = ax3.bar(range(len(capabilities)), cap_scores,
                   color=[presentation_colors['primary'], presentation_colors['secondary'],
                          presentation_colors['success'], presentation_colors['accent'],
                          presentation_colors['warning']], alpha=0.8)

    ax3.set_title('System Capabilities', fontsize=16, fontweight='bold')
    ax3.set_ylabel('Effectiveness Score')
    ax3.set_xticks(range(len(capabilities)))
    ax3.set_xticklabels(capabilities, rotation=45, ha='right')
    ax3.set_ylim(0, 100)

    # Add value labels on bars
    for bar, score in zip(bars, cap_scores):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width() / 2., height + 1,
                 f'{score}%', ha='center', va='bottom', fontweight='bold')

    # ROI Projection (Bottom Right)
    months = ['Month 1', 'Month 6', 'Month 12', 'Month 18', 'Month 24']
    roi_values = [-20, 15, 45, 85, 120]  # ROI percentage

    ax4.plot(months, roi_values, marker='o', linewidth=3, markersize=8,
             color=presentation_colors['success'])
    ax4.fill_between(months, roi_values, alpha=0.3, color=presentation_colors['success'])
    ax4.axhline(y=0, color='black', linestyle='--', alpha=0.5)
    ax4.set_title('ROI Projection', fontsize=16, fontweight='bold')
    ax4.set_ylabel('ROI (%)')
    ax4.grid(True, alpha=0.3)

    # Highlight break-even point
    ax4.annotate('Break-even Point', xy=('Month 6', 15), xytext=('Month 6', 50),
                 arrowprops=dict(arrowstyle='->', color=presentation_colors['accent'], lw=2),
                 fontsize=12, ha='center', color=presentation_colors['accent'], fontweight='bold')

    plt.tight_layout()
    plt.show()

    return fig


executive_summary = create_executive_summary()


# System Architecture Overview for Presentation
def create_architecture_overview():
    fig, ax = plt.subplots(1, 1, figsize=(16, 10))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 8)
    ax.axis('off')

    # Title
    ax.text(5, 7.5, 'Smart Parking System Architecture',
            ha='center', va='center', fontsize=20, fontweight='bold')

    # Main components with professional styling
    components = [
        {'name': 'Entry\nSystem', 'pos': (1.5, 6), 'color': presentation_colors['primary']},
        {'name': 'Payment\nProcessor', 'pos': (5, 6), 'color': presentation_colors['secondary']},
        {'name': 'Exit\nSystem', 'pos': (8.5, 6), 'color': presentation_colors['accent']},

        {'name': 'Redis\nCache', 'pos': (2.5, 4), 'color': presentation_colors['warning']},
        {'name': 'SQLite\nDatabase', 'pos': (7.5, 4), 'color': presentation_colors['success']},

        {'name': 'AI Vision\n(YOLO + OCR)', 'pos': (1.5, 2), 'color': presentation_colors['dark']},
        {'name': 'Arduino\nControllers', 'pos': (5, 2), 'color': presentation_colors['primary']},
        {'name': 'Real-time\nDashboard', 'pos': (8.5, 2), 'color': presentation_colors['secondary']},
    ]

    # Draw components
    for comp in components:
        circle = plt.Circle(comp['pos'], 0.6, facecolor=comp['color'],
                            edgecolor='white', linewidth=3, alpha=0.8)
        ax.add_patch(circle)
        ax.text(comp['pos'][0], comp['pos'][1], comp['name'],
                ha='center', va='center', fontsize=11, fontweight='bold', color='white')

    # Add connecting arrows
    connections = [
        ((1.5, 5.4), (2.5, 4.6)),  # Entry to Redis
        ((5, 5.4), (2.5, 4.6)),  # Payment to Redis
        ((8.5, 5.4), (7.5, 4.6)),  # Exit to SQLite
        ((3.1, 4), (6.9, 4)),  # Redis to SQLite
        ((5, 5.4), (5, 2.6)),  # Payment to Arduino
        ((1.5, 5.4), (1.5, 2.6)),  # Entry to AI
        ((8.5, 5.4), (8.5, 2.6)),  # Exit to Dashboard
    ]

    for start, end in connections:
        arrow = patches.FancyArrowPatch(start, end,
                                        arrowstyle='->', mutation_scale=20,
                                        color='gray', linewidth=2, alpha=0.7)
        ax.add_patch(arrow)

    # Add performance indicators
    performance_texts = [
        {'text': '< 2sec\nresponse', 'pos': (1.5, 0.5), 'color': presentation_colors['success']},
        {'text': '99.2%\naccuracy', 'pos': (5, 0.5), 'color': presentation_colors['success']},
        {'text': 'Real-time\nmonitoring', 'pos': (8.5, 0.5), 'color': presentation_colors['success']},
    ]

    for perf in performance_texts:
        ax.text(perf['pos'][0], perf['pos'][1], perf['text'],
                ha='center', va='center', fontsize=10, color=perf['color'],
                fontweight='bold', bbox=dict(boxstyle="round,pad=0.3",
                                             facecolor='white', edgecolor=perf['color']))

    plt.tight_layout()
    plt.show()

    return fig


architecture_overview = create_architecture_overview()


# Live Demo Instructions Card
def create_demo_instructions():
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 8)
    ax.axis('off')

    # Background
    bg_rect = patches.Rectangle((0.5, 0.5), 9, 7, linewidth=2,
                                edgecolor=presentation_colors['primary'],
                                facecolor=presentation_colors['light'], alpha=0.3)
    ax.add_patch(bg_rect)

    # Title
    ax.text(5, 7.2, 'Live Demonstration Guide',
            ha='center', va='center', fontsize=18, fontweight='bold',
            color=presentation_colors['primary'])

    # Instructions
    instructions = [
        "1. ðŸ“± Scan QR code to access live dashboard",
        "2. ðŸš— Watch real-time vehicle detection",
        "3. ðŸ’³ See automated payment processing",
        "4. ðŸ“Š View live analytics and statistics",
        "5. ðŸ”’ Observe security alert system"
    ]

    for i, instruction in enumerate(instructions):
        ax.text(1, 6 - i * 0.8, instruction, fontsize=14,
                color=presentation_colors['dark'], va='center')

        # Add colored bullet point
        circle = plt.Circle((0.7, 6 - i * 0.8), 0.1,
                            color=presentation_colors['secondary'], alpha=0.8)
        ax.add_patch(circle)

    # Add QR code placeholder
    qr_rect = patches.Rectangle((7, 2), 2, 2, linewidth=2,
                                edgecolor=presentation_colors['primary'],
                                facecolor='white')
    ax.add_patch(qr_rect)
    ax.text(8, 3, 'QR CODE\nHERE', ha='center', va='center',
            fontsize=12, fontweight='bold', color=presentation_colors['primary'])

    # Add footer
    ax.text(5, 0.8, 'Experience the future of parking management in real-time!',
            ha='center', va='center', fontsize=12, style='italic',
            color=presentation_colors['accent'])

    plt.tight_layout()
    plt.show()

    return fig


demo_instructions = create_demo_instructions()


# Benefits and ROI Summary
def create_benefits_summary():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

    # Benefits comparison (Before vs After)
    categories = ['Processing\nTime', 'Accuracy', 'Labor\nCosts', 'User\nSatisfaction', 'Revenue']
    before_values = [120, 75, 100, 60, 80]  # Relative scale
    after_values = [15, 99, 40, 95, 140]  # Relative scale

    x = np.arange(len(categories))
    width = 0.35

    bars1 = ax1.bar(x - width / 2, before_values, width, label='Before Implementation',
                    color=presentation_colors['accent'], alpha=0.7)
    bars2 = ax1.bar(x + width / 2, after_values, width, label='After Implementation',
                    color=presentation_colors['success'], alpha=0.7)

    ax1.set_title('Performance Improvement', fontsize=16, fontweight='bold')
    ax1.set_ylabel('Performance Index')
    ax1.set_xticks(x)
    ax1.set_xticklabels(categories)
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Add improvement percentages
    for i, (before, after) in enumerate(zip(before_values, after_values)):
        improvement = ((after - before) / before) * 100
        color = presentation_colors['success'] if improvement > 0 else presentation_colors['accent']
        ax1.text(i, max(before, after) + 5, f'{improvement:+.0f}%',
                 ha='center', va='bottom', fontweight='bold', color=color)

    # Cost-Benefit Analysis
    years = ['Year 0', 'Year 1', 'Year 2', 'Year 3']
    investment = [-50000, -10000, -5000, -5000]  # Initial investment and maintenance
    savings = [0, 35000, 45000, 55000]  # Annual savings
    net_benefit = [inv + sav for inv, sav in zip(investment, savings)]

    ax2.bar(years, investment, label='Investment', color=presentation_colors['accent'], alpha=0.7)
    ax2.bar(years, savings, bottom=investment, label='Savings',
            color=presentation_colors['success'], alpha=0.7)
    ax2.plot(years, net_benefit, marker='o', linewidth=3, markersize=8,
             color=presentation_colors['primary'], label='Net Benefit')

    ax2.set_title('Financial Impact Analysis', fontsize=16, fontweight='bold')
    ax2.set_ylabel('Amount ($)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='black', linestyle='--', alpha=0.5)

    # Add cumulative savings annotation
    cumulative_savings = sum(savings) + sum(investment)
    ax2.annotate(f'Total ROI: ${cumulative_savings:,}',
                 xy=('Year 3', net_benefit[-1]),
                 xytext=('Year 2', net_benefit[-1] + 20000),
                 arrowprops=dict(arrowstyle='->', color=presentation_colors['success'], lw=2),
                 fontsize=12, ha='center', color=presentation_colors['success'],
                 fontweight='bold')

    plt.tight_layout()
    plt.show()

    return fig


benefits_summary = create_benefits_summary()


# Technology Innovation Showcase
def create_innovation_showcase():
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Innovation & Technology Leadership', fontsize=18, fontweight='bold')

    # AI/ML Capabilities
    ai_features = ['License Plate\nRecognition', 'Vehicle\nDetection', 'Pattern\nAnalysis',
                   'Predictive\nAnalytics', 'Anomaly\nDetection']
    ai_accuracy = [99.2, 98.5, 96.8, 94.2, 97.1]

    colors = plt.cm.viridis(np.linspace(0, 1, len(ai_features)))
    bars = ax1.barh(ai_features, ai_accuracy, color=colors, alpha=0.8)
    ax1.set_title('AI/ML Capabilities', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Accuracy (%)')
    ax1.set_xlim(90, 100)

    # Add accuracy labels
    for bar, acc in zip(bars, ai_accuracy):
        width = bar.get_width()
        ax1.text(width + 0.1, bar.get_y() + bar.get_height() / 2,
                 f'{acc}%', ha='left', va='center', fontweight='bold')

    # Real-time Performance Metrics
    metrics = ['Response Time', 'Throughput', 'Reliability', 'Scalability']
    current_performance = [95, 88, 99, 85]
    industry_average = [60, 65, 80, 70]

    x = np.arange(len(metrics))
    width = 0.35

    ax2.bar(x - width / 2, industry_average, width, label='Industry Average',
            color=presentation_colors['light'], edgecolor=presentation_colors['dark'])
    ax2.bar(x + width / 2, current_performance, width, label='Our System',
            color=presentation_colors['secondary'], alpha=0.8)

    ax2.set_title('Performance vs Industry', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Performance Score')
    ax2.set_xticks(x)
    ax2.set_xticklabels(metrics, rotation=45, ha='right')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # Integration Complexity
    systems = ['Hardware', 'Software', 'AI Models', 'Database', 'Analytics', 'Security']
    integration_levels = [85, 95, 90, 88, 92, 87]

    # Create a radar chart
    angles = np.linspace(0, 2 * np.pi, len(systems), endpoint=False).tolist()
    integration_levels += integration_levels[:1]  # Complete the circle
    angles += angles[:1]

    ax3.plot(angles, integration_levels, 'o-', linewidth=3,
             color=presentation_colors['accent'], markersize=8)
    ax3.fill(angles, integration_levels, alpha=0.25, color=presentation_colors['accent'])
    ax3.set_xticks(angles[:-1])
    ax3.set_xticklabels(systems)
    ax3.set_ylim(0, 100)
    ax3.set_title('System Integration Maturity', fontsize=14, fontweight='bold')
    ax3.grid(True)

    # Future Roadmap
    quarters = ['Q1 2024', 'Q2 2024', 'Q3 2024', 'Q4 2024', 'Q1 2025']
    features = ['Mobile App', 'IoT Sensors', 'Predictive Analytics', 'Cloud Migration', 'AI Enhancement']

    for i, (quarter, feature) in enumerate(zip(quarters, features)):
        color = presentation_colors['success'] if i < 2 else presentation_colors['warning'] if i < 4 else \
        presentation_colors['secondary']
        ax4.barh(i, 1, color=color, alpha=0.7)
        ax4.text(0.5, i, f'{quarter}: {feature}', ha='center', va='center',
                 fontweight='bold', color='white')

    ax4.set_title('Innovation Roadmap', fontsize=14, fontweight='bold')
    ax4.set_xlim(0, 1)
    ax4.set_yticks(range(len(quarters)))
    ax4.set_yticklabels([])
    ax4.set_xticks([])

    plt.tight_layout()
    plt.show()

    return fig


innovation_showcase = create_innovation_showcase()


# Presentation Summary & Call to Action
def create_presentation_summary():
    fig, ax = plt.subplots(1, 1, figsize=(14, 10))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis('off')

    # Main title
    ax.text(5, 9.2, 'Smart Parking Management System',
            ha='center', va='center', fontsize=24, fontweight='bold',
            color=presentation_colors['primary'])

    ax.text(5, 8.7, 'Revolutionizing Parking Through AI & Automation',
            ha='center', va='center', fontsize=16, style='italic',
            color=presentation_colors['secondary'])

    # Key achievements
    achievements = [
        "âœ… 99.2% License Plate Recognition Accuracy",
        "âœ… < 2 Second Processing Time",
        "âœ… 40% Operational Cost Reduction",
        "âœ… Real-time Analytics & Monitoring",
        "âœ… Automated Payment Processing",
        "âœ… Advanced Security Features"
    ]

    for i, achievement in enumerate(achievements):
        ax.text(1, 7.5 - i * 0.4, achievement, fontsize=12,
                color=presentation_colors['success'], fontweight='bold', va='center')

    # Technology highlights
    tech_box = patches.FancyBboxPatch((5.5, 4.5), 4, 3,
                                      boxstyle="round,pad=0.2",
                                      facecolor=presentation_colors['light'],
                                      edgecolor=presentation_colors['primary'],
                                      linewidth=2, alpha=0.8)
    ax.add_patch(tech_box)

    ax.text(7.5, 7.2, 'Technology Stack', ha='center', va='center',
            fontsize=14, fontweight='bold', color=presentation_colors['primary'])

    tech_stack = [
        "ðŸ¤– AI: YOLO + Tesseract OCR",
        "âš¡ Cache: Redis (< 1ms)",
        "ðŸ’¾ Database: SQLite",
        "ðŸ”§ Hardware: Arduino",
        "ðŸ“Š Frontend: Real-time Dashboard",
        "ðŸ Backend: Python Flask"
    ]

    for i, tech in enumerate(tech_stack):
        ax.text(5.7, 6.8 - i * 0.25, tech, fontsize=10,
                color=presentation_colors['dark'], va='center')

    # Call to action
    cta_box = patches.FancyBboxPatch((1, 1), 8, 2.5,
                                     boxstyle="round,pad=0.3",
                                     facecolor=presentation_colors['accent'],
                                     alpha=0.9)
    ax.add_patch(cta_box)

    ax.text(5, 2.8, 'ðŸš€ Ready for Implementation',
            ha='center', va='center', fontsize=18, fontweight='bold', color='white')

    ax.text(5, 2.3, 'Scan QR Code for Live Demo',
            ha='center', va='center', fontsize=14, color='white')

    ax.text(5, 1.8, 'Experience the Future of Parking Management',
            ha='center', va='center', fontsize=12, style='italic', color='white')

    # Contact info
    ax.text(5, 0.5, 'Questions? Contact: [Your Contact Information]',
            ha='center', va='center', fontsize=10,
            color=presentation_colors['dark'], style='italic')

    plt.tight_layout()
    plt.show()

    return fig


presentation_summary = create_presentation_summary()

print("\nðŸŽ‰ Presentation toolkit complete!")
print("ðŸ“ All assets generated and ready for your presentation")
print("ðŸ“± QR code ready for audience interaction")
print("ðŸ“Š Professional visuals created for executive presentation")
```
